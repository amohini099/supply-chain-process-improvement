```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.graphics.gofplots as sm
```

```python
np.random.seed(0)
```

## Normal Distribution

```python
# Parameters
mu = 23
sigma = 7
plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title('Normal Distribution N(23,7)')
plt.show()
```

### Probability n>30 items

```python
# Parameters
mu = 23
sigma = 7
# Set up probability function
stats.norm(mu, sigma)
# Probability X = 30
x1 = 30
# Cumulative Density Function
y1 = stats.norm(mu, sigma).cdf(x1)
# Probability Density Function
z1 = stats.norm(mu, sigma).pdf(x1)

plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.axvline(x1, color='red')
plt.axhline(z1, color='red')
plt.title('Normal Distribution N(23,7)')
print("Probability to have less than {} items/carton is {}%".format(x1, round(100*y1,2)))
plt.show()
```

### Probability = 95%

```python
# Parameters
mu = 23
sigma = 7
# Set up probability function
stats.norm(mu, sigma)
# Probability X = 30
y1 = 0.95
# Percent point function (inverse of cdf)
x1 = stats.norm(mu, sigma).ppf(y1)
# Probability Density Function
z1 = stats.norm(mu, sigma).pdf(x1)

plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.axvline(x1, color='red')
plt.axhline(z1, color='red')
plt.title('Normal Distribution N(23,7)')
print("We have 95% of probability that X <= {} items/carton".format(round(x1,3)))
plt.show()
```
